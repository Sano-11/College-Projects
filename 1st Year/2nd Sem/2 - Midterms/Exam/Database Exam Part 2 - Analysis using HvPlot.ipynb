## Setup — Import Libraries and Load Dataset

```python
import os
import pandas as pd
import hvplot.pandas  # this gives us the .hvplot accessor on DataFrames

# Load the dataset (make sure the CSV is in the same folder as this notebook)
script_dir = os.path.dirname(os.path.abspath('__file__'))
filename = os.path.join(script_dir, "ai_job_replacement_2020_2026_v2.csv")

df = pd.read_csv(filename)

print("Dataset loaded successfully!")
print(f"Shape: {df.shape[0]} rows x {df.shape[1]} columns")
print()
print("Column names:")
print(list(df.columns))
```

```python
# Summary statistics
df.describe()
```

## Question 1: What is the average automation risk percentage per industry?

We group by `industry` and compute the mean of `automation_risk_percent`,
then sort from highest to lowest.

```python
# Group by industry and get the average automation risk
q1 = (
    df.groupby("industry")["automation_risk_percent"]
    .mean()
    .round(2)
    .sort_values(ascending=False)
    .reset_index()
)
q1.columns = ["Industry", "Avg Automation Risk (%)"]
q1
```

### Chart 1 — Bar Chart: Average Automation Risk % by Industry

```python
q1.hvplot.bar(
    x="Industry",
    y="Avg Automation Risk (%)",
    title="Average Automation Risk % by Industry",
    xlabel="Industry",
    ylabel="Avg Automation Risk (%)",
    color="darkred",
    rot=30,
    height=400,
    width=700
)
```

---
## Question 2: Which countries have the highest average AI replacement score?

We select the `country` and `ai_replacement_score` columns, group by country,
then sort to find which countries are most affected.

```python
# Select only the columns we need, group by country, and sort
q2 = (
    df[["country", "ai_replacement_score"]]
    .groupby("country")["ai_replacement_score"]
    .mean()
    .round(2)
    .sort_values(ascending=False)
    .reset_index()
)
q2.columns = ["Country", "Avg AI Replacement Score"]
q2
```

### Chart 2 — Horizontal Bar Chart: AI Replacement Score by Country

```python
q2.hvplot.barh(
    x="Country",
    y="Avg AI Replacement Score",
    title="Average AI Replacement Score by Country",
    xlabel="Avg AI Replacement Score",
    ylabel="Country",
    color="maroon",
    height=400,
    width=700
)
```

---
## Question 3: How does the automation risk category affect the average salary change?

We group by `automation_risk_category` and look at the average `salary_change_percent`
to see if high-risk jobs lose more salary.

```python
# Group by risk category, compute average salary change
q3 = (
    df.groupby("automation_risk_category")["salary_change_percent"]
    .agg(["mean", "median", "count"])
    .round(2)
    .reset_index()
)
q3.columns = ["Risk Category", "Mean Salary Change (%)", "Median Salary Change (%)", "Count"]
q3
```

### Chart 3 — Bar Chart: Salary Change % by Automation Risk Category

```python
q3.hvplot.bar(
    x="Risk Category",
    y="Mean Salary Change (%)",
    title="Average Salary Change % by Automation Risk Category",
    xlabel="Automation Risk Category",
    ylabel="Mean Salary Change (%)",
    color=["green", "orange", "red"],
    height=400,
    width=600
)
```

---
## Question 4: How have AI adoption levels and reskilling urgency changed over the years?

We group by `year` and track the average `ai_adoption_level` and `reskilling_urgency_score`
to observe trends from 2020 to 2026.

```python
# Group by year and get averages
q4 = (
    df.groupby("year")[["ai_adoption_level", "reskilling_urgency_score"]]
    .mean()
    .round(2)
    .reset_index()
)
q4.columns = ["Year", "Avg AI Adoption Level", "Avg Reskilling Urgency Score"]
q4
```

### Chart 4 — Line Chart: AI Adoption & Reskilling Urgency Over Time

```python
# Plot both metrics as lines on the same chart
chart_adoption = q4.hvplot.line(
    x="Year",
    y="Avg AI Adoption Level",
    label="AI Adoption Level",
    color="darkred",
    line_width=2,
    markers=True,
    height=400,
    width=700
)

chart_reskilling = q4.hvplot.line(
    x="Year",
    y="Avg Reskilling Urgency Score",
    label="Reskilling Urgency Score",
    color="goldenrod",
    line_width=2,
    markers=True
)

# Overlay both lines on one chart
(chart_adoption * chart_reskilling).opts(
    title="AI Adoption Level vs Reskilling Urgency Score (2020-2026)",
    legend_position="top_left"
)
```

## Question 5: What are the top 10 job roles with the highest average salary before AI disruption?

We select `job_role` and `salary_before_usd`, group by job role, sort descending,
and take the top 10.

```python
# Select columns, group, sort, and get top 10
q5 = (
    df[["job_role", "salary_before_usd"]]
    .groupby("job_role")["salary_before_usd"]
    .mean()
    .round(2)
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
q5.columns = ["Job Role", "Avg Salary Before AI (USD)"]
q5
```

### Chart 5 — Horizontal Bar Chart: Top 10 Highest-Paying Job Roles

```python
q5.hvplot.barh(
    x="Job Role",
    y="Avg Salary Before AI (USD)",
    title="Top 10 Job Roles by Average Salary Before AI",
    xlabel="Avg Salary Before AI (USD)",
    ylabel="Job Role",
    color="maroon",
    height=450,
    width=750
)
```

## Bonus Question: Is there a relationship between automation risk and salary change?

We use a scatter plot to visualize the relationship between `automation_risk_percent`
and `salary_change_percent`, colored by industry.

```python
# Scatter plot: automation risk vs salary change, colored by industry
df.hvplot.scatter(
    x="automation_risk_percent",
    y="salary_change_percent",
    by="industry",
    title="Automation Risk % vs Salary Change % (by Industry)",
    xlabel="Automation Risk %",
    ylabel="Salary Change %",
    alpha=0.4,
    height=450,
    width=750
)
```

## Bonus Question: What is the distribution of AI replacement scores across all job roles?

A histogram shows how the AI replacement scores are distributed across the dataset.

```python
# Histogram of AI replacement scores
df.hvplot.hist(
    y="ai_replacement_score",
    bins=30,
    title="Distribution of AI Replacement Scores",
    xlabel="AI Replacement Score",
    ylabel="Count",
    color="darkred",
    alpha=0.8,
    height=400,
    width=700
)
```
