# Data Analysis with Pandas Activity
This notebook contains the analysis of global suicide data.

```python
import pandas as pd 
import os
```

## Activity 1 (Country, Year, Age, Sex, Suicide)

```python

file_path = "suicide.csv"

# Read the CSV
df = pd.read_csv(file_path)
print(df[['country', 'year', 'sex', 'age', 'suicides/100k pop']])
print(df[['country', 'year', 'sex', 'age', 'suicides/100k pop']])
```

## Activity 2 Suicide record. Display year, sex, age, suicides_n, and suicides/100k pop columns.

```python
ph_records = df[df['country'] == 'Philippines']
display(ph_records[['year', 'sex', 'age', 'suicides_no', 'suicides/100k pop']])

ph_2011 = df[(df['country'] == 'Philippines') & (df['year'] == 2011)]
ph_2011[['year', 'sex', 'age', 'suicides_no', 'suicides/100k pop']]    
```

## Activity 3 Suicide record in the year 2011. Display year, sex, age, suicides_n, and suicides/100k pop columns

```python
ph_2011 = df[(df['country'] == 'Philippines') & (df['year'] == 2011)]
display(ph_2011[['year', 'sex', 'age', 'suicides_no', 'suicides/100k pop']])
```

## Activity 4 Country with the highest suicide per 100k population in 2005.Display year, sex, age, and suicides/100k pop columns.

```python
highest_2005 = df[df['year'] == 2005].sort_values(by='suicides/100k pop', ascending=False).head(1)
display(highest_2005[['country', 'year', 'sex', 'age', 'suicides/100k pop']])
```

## Activity 5 Total suicide cases (suicides_n) and suicides/100k pop per year. Sort in descending order based on suicides_n.


```python
year_stats = df.groupby('year').agg({'suicides_no': 'sum', 'suicides/100k pop': 'sum'})
display(year_stats.sort_values(by='suicides_no', ascending=False))
```

## Activity 6 total suicide cases (suicides_n) and suicides/100k pop per gender. Sort in descending order based on suicides_n.

```python
gender_stats = df.groupby('sex').agg({'suicides_no': 'sum', 'suicides/100k pop': 'sum'})
display(gender_stats.sort_values(by='suicides_no', ascending=False))
```

## Activity 7 Total suicide cases (suicides_n) and suicides/100k pop per age. Sort in descending order based on suicides_n.


```python
age_stats = df.groupby('age').agg({'suicides_no': 'sum', 'suicides/100k pop': 'sum'})
display(age_stats.sort_values(by='suicides_no', ascending=False))
```

## Activity 8 Total suicide cases (suicides_n) per country and per year. Sort in descending order based on suicides_n.

```python
country_year_stats = df.groupby(['country', 'year'])[['suicides_no']].sum()
display(country_year_stats.sort_values(by='suicides_no', ascending=False))
```

## Activity 9 Total suicide cases in the Philippines based on year, gender, and age.

```python
ph_total_stats = df[df['country'] == 'Philippines'].groupby(['year', 'sex', 'age'])[['suicides_no']].sum()
display(ph_total_stats)
```

## Activity 10 year and number of suicide cases that the Philippines has the highest number of combined suicide cases.

```python
ph_highest_year = df[df['country'] == 'Philippines'].groupby('year')[['suicides_no']].sum()
display(ph_highest_year.sort_values(by='suicides_no', ascending=False).head(1))
```
