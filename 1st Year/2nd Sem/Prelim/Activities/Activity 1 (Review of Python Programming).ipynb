# <font color ='green'> Activity 1 : Peso Calculator

```python
# Activity 1
# Write a program that will accept input for the number of 25 centavos,
# 1 peso, 5 pesos, 10 pesos, 20 pesos, 50 pesos, 100 pesos, 200 pesos
# 500 pesos, and 1000 pesos
# Compute and display the total amount of money

money = {
    "25 centavos": 0.25,
    "1 peso coin": 1,
    "5 peso coin": 5,
    "10 peso coin": 10,
    "20 peso bill": 20,
    "50 peso bill": 50,
    "100 peso bill": 100,
    "200 peso bill": 200,
    "500 peso bill": 500,
    "1000 peso bill": 1000
}
# Collect inputs and compute total
total = 0
for name, value in money.items():
    count = int(input(f"Enter number of {name}: "))
    total += count * value

# Display result
print(f"\nTotal amount of money: ₱{total:,.2f}")
```

# <font color ='green'> Activity 2 : Grades Calculator

```python
# Activity 2
# Create a program that will accept Prelim, Midterm, Semi-Final, Final, Quiz, and Project Grade.
# Compute the final grade: FG = (prelime x 15%) + (midterm x 15%) + (semifinal x 15%) + (final x 15%) + (quiz x 25%) + (project x 15%)
# Display a remark Congratulations! You passed the course if final grade is 75 and above; otherwise display Sorry, you failed

weights = {
    "Prelim": 0.15,
    "Midterm": 0.15,
    "Semifinal": 0.15,
    "Final": 0.15,
    "Quiz": 0.25,
    "Project": 0.15}
grades = {}
for subject, weight in weights.items():
    grades[subject] = int(input(f"Enter {subject} Grade: "))
    
Final_Grade = sum(grades[subject] * weight for subject, weight in weights.items())

if Final_Grade > 100:
    print("\nInvalid")
elif Final_Grade >= 75:
    print(f"\nFinal Grade: {Final_Grade:.2f}")
    print("Congratulations! You passed the course.")
else:
    print(f"\nFinal Grade: {Final_Grade:.2f}")
    print("Sorry, you failed.")
```

# <font color = 'green'> Activity 4 : Odd or Even?

```python
# Activity 4
# Accept 10 Integer numbers.
# Count and display how many odd and even number input.

numbers = [int(input(f"Enter number {i+1}: ")) for i in range(10)]

evencount = sum(1 for n in numbers if n % 2 == 0)
oddcount = len(numbers) - evencount

print("\nEven numbers:", evencount)
print("Odd numbers:", oddcount)
```

# <font color = 'green'> Activity 5 : Prime or Composite?

```python
# Activity 5
# Input a number and display if the number is a Prime Number or Not a Prime Number

num = int(input("Enter a Number: "))

if num < 2:
    print(f"{num} is NOT a Prime Number")
elif all(num % i != 0 for i in range(2, int(num**0.5) + 1)):
    print(f"{num} is a Prime Number")
else:
    print(f"{num} is NOT a Prime Number")
```

# <font color = 'green'> Activity 6 : Series of Prime Numbers

```python
# Activity 6
# Display a series of prime numbers based on user input.

num = int(input("Enter how many prime numbers to display: "))
count, candidate = 0, 2

while count < num:
    if all(candidate % i != 0 for i in range(2, int(candidate**0.5) + 1)):
        print(candidate, end=" ")
        count += 1
    candidate += 1
```

# <font color = 'green'> Activity 7 : Nth Term of Fibonacci Series

```python
# Activity 7
# Create a program that will ask user to input the nth Fibonacci number series.
# Compute and display the Fibonacci number based on the input

n = int(input("Enter the number of terms in Fibonacci series: "))

a, b = 0, 1
print("Fibonacci series:")

for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

# <font color = 'green'> Activity 8  : List Content, Size, List, Highest, Lowest

```python
# Activity 8
# Accept integer number and store it to a list until user input is 0.
# Display the content of the list and size of the list.
# Display the highest and lowest numbers.

numbers = []

while True:
    n = int(input("Enter a number (0 to stop): "))
    if n == 0:
        break
    numbers.append(n)

print("\nNumber/s entered:", numbers)
print("Size of list:", len(numbers))
print("Highest number:", max(numbers) if numbers else "None")
print("Lowest number:", min(numbers) if numbers else "None")
```

# <font color = 'green'> Activity 9  : List Content, Size, List, Highest, Lowest, Index

```python
# Activity 9
# Accept integer number and store it to a list until user input is 0.
# Display the content of the list and size of the list.
# Display the highest and lowest numbers in the list and the index number it can be found

numbers = []

while True:
    n = int(input("Enter a number (0 to stop): "))
    if n == 0:
        break
    numbers.append(n)

print("\nNumbers entered:", numbers)
print("Size of list:", len(numbers))

if numbers:
    highest = max(numbers)
    lowest = min(numbers)
    print("Highest number:", highest, "at index", numbers.index(highest))
    print("Lowest number:", lowest, "at index", numbers.index(lowest))
else:
    print("No numbers were entered.")
```

# <font color = 'green'> Activity 10  : Student Name and Grade Dictionary

```python
# Activity 10
# Accept student name and grade and save it to a dictionary.
# Display the content of the dictionary.
# Display the name  and grade of student with the highest and lowest grade.
students = {}

while True:
    name = input("Enter student name (or 'stop' to finish): ")
    if name.lower() == "stop":
        break
    grade = int(input(f"Enter grade for {name}: "))
    students[name] = grade

print("\nStudent Records:")
for name, grade in students.items():
    print(f"{name}: {grade}")

if students:
    highest = max(students, key=students.get)
    lowest = min(students, key=students.get)
    print(f"\nHighest Grade: {highest} - {students[highest]}")
    print(f"Lowest Grade: {lowest} - {students[lowest]}")
else:
    print("\nNo student records entered.")
```

# <font color = 'green'> Activity 12  : String Vowel and Consonant counter, Reverse Order

```python
# Activity 12
# Accept string input.
# Count vowel and consonant characters in the string.
# Display the string in reverse order.

text = input("Enter a string: ")

vowels = "aeiouAEIOU"
vowel_count = sum(1 for ch in text if ch in vowels and ch.isalpha())
consonant_count = sum(1 for ch in text if ch.isalpha() and ch not in vowels)

print(f"\nVowels: {vowel_count}")
print(f"Consonants: {consonant_count}")
print("Reversed string:", text[::-1])
```
